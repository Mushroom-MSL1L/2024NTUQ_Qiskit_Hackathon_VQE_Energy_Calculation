# import package

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# load  files 

```python
# read xlsx
file_path = 'VQE data.xlsx'
# first sheet
df_first_sheet = pd.read_excel(file_path, sheet_name=0)
# second sheet
df_second_sheet = pd.read_excel(file_path, sheet_name=1)
# third sheet
df_third_sheet = pd.read_excel(file_path, sheet_name=2)
# third sheet
df_four_sheet = pd.read_excel(file_path, sheet_name=3)
# third sheet
df_five_sheet = pd.read_excel(file_path, sheet_name=4)
```

# picture settings

```python
plt.rcParams['font.family'] = "Times New Roman"  # UbuntuでのTimes New Romanの代用
plt.rcParams['xtick.direction'] = "in"
plt.rcParams['ytick.direction'] = "in"
plt.rcParams['xtick.major.width'] = 1.0
plt.rcParams['ytick.major.width'] = 1.0
plt.rcParams['xtick.minor.width'] = 0.5
plt.rcParams['xtick.major.size'] = 5
plt.rcParams['ytick.major.size'] = 5
plt.rcParams['xtick.minor.size'] = 2
plt.rcParams['font.size'] = 20
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.linewidth'] = 1.0
```

# VQE data 1 -- without taper
Ansatz : HEA <br>
Sampling Method : bfgs <br>
Electron : 10 <br>
Orbit : 8 <br>
Z2 : False <br>
Runtime : 43 minutes <br>
Final Value : -1537.934621 <br>

```python
# vqe data 1
# get true y val
true_val = df_first_sheet.iloc[0, 0]
# get sim val
sim_val = df_third_sheet.iloc[:, 0]
# caculate x val (index*33 + 1)
shots = sim_val.index * 33 + 1

# set text message
textstr = (
    f"Ansatz : HEA\n"
    f"Sampling Method : bfgs\n"
    f"Electron : 10\n"
    f"Orbit : 8\n"
    f"Z2 : False\n"
    f"Runtime : 43 minutes\n"
    f"Final Value : -1537.934621"
)

# whole pic
fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(shots, sim_val.values, marker='o',
         linestyle='-', color='b', label='Simulate')
ax1.axhline(y=true_val, color='r', linestyle='--',
            label=f'Theoretical Value = {true_val}')
ax1.set_xlabel('Shots')
ax1.set_ylabel('Ground Energy(Hartree)')
ax1.legend()
ax1.grid(True)
fig1.suptitle(
    r'$\mathrm{LiCoO_2}$ Simulation without Qubit tapering', fontsize=30)
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
fig1.text(0.58, 0.7, textstr, fontsize=20, verticalalignment='top',
          horizontalalignment='left', bbox=props)
# save to pdf
fig1.savefig('licoo2_sim_without_taper_whole.pdf',
             dpi=300, bbox_inches='tight')

# specific pic
fig2, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(shots, sim_val.values, marker='o',
         linestyle='-', color='b', label='Simulate')
ax2.axhline(y=true_val, color='r', linestyle='--',
            label=f'Y Value = {true_val}')
ax2.set_xlabel('Shots')
ax2.set_ylim(-1538, -1537)
ax2.set_ylabel('Ground Energy(Hartree)')
ax2.legend()
ax2.grid(True)
fig2.suptitle(
    r'$\mathrm{LiCoO_2}$ Simulation without Qubit tapering', fontsize=30)
fig2.text(0.58, 0.7, textstr, fontsize=20, verticalalignment='top',
          horizontalalignment='left', bbox=props)
# save to pdf
fig2.savefig('licoo2_sim_without_taper_specific.pdf',
             dpi=300, bbox_inches='tight')

# show
plt.show()
```
